# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type (used city_data instead of pyber to avoid merged issues as no cleaning has been done)
sum_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
sum_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_by_type
```

```python
#setting df separated by city type urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
urban_cities_df.tail()
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare__by_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare__by_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_type = total_fare_by_type/sum_drivers_by_type
# avg_fare_per_driver_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]/pyber_data_df.groupby(["type"]).sum()["driver_count"]
avg_fare_per_driver_by_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'type': ['Urban', 'Suburban', 'Rural'],
                                 'Total Rides': [total_rides_by_type["Urban"], total_rides_by_type["Suburban"], total_rides_by_type["Rural"]],
                                 'Total Drivers': [sum_drivers_by_type["Urban"], sum_drivers_by_type["Suburban"], sum_drivers_by_type["Rural"]],
                                 'Total Fares': [total_fare_by_type["Urban"], total_fare_by_type["Suburban"], total_fare_by_type["Rural"]],
                                 'Average Fare per Ride': [avg_fare__by_type["Urban"], avg_fare__by_type["Suburban"], avg_fare__by_type["Rural"]],
                                 'Average Fare per Driver': [avg_fare_per_driver_by_type["Urban"], avg_fare_per_driver_by_type["Suburban"], avg_fare_per_driver_by_type["Rural"]]})
    
pyber_summary_df = pyber_summary_df.set_index('type')
    
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# trimmed decimals to hundreths to represent USD($)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_per_type_and_date = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_fares_per_type_and_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_per_type_and_date = sum_fares_per_type_and_date.reset_index()
sum_fares_per_type_and_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fares_per_type_and_date_piv = pyber_data_df.pivot(index='date',columns='type',values='fare')
sum_fares_per_type_and_date_piv
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_of_fares_2019_01to04_df = sum_fares_per_type_and_date_piv.loc["2019-01-01":"2019-04-29"]
sum_of_fares_2019_01to04_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
sum_of_fares_2019_01to04_df.index = pd.to_datetime(sum_of_fares_2019_01to04_df.index)
sum_of_fares_2019_01to04_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_of_fares_2019_01to04_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_by_week_2019_01to04_df = sum_of_fares_2019_01to04_df.resample('W').sum()
sum_of_fares_by_week_2019_01to04_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

sum_of_fares_by_week_2019_01to04_df.plot(
    use_index=True,
    xlabel='By Week',
    ylabel='Fare ($USD)',
    grid=True,
    legend=True,
    figsize=(15,5),
    fontsize=14,
    title='Total Fare by City Type'
)
plt.legend(loc='lower left')
# plt.savefig("Challenge/PyBer_fare_summary.png") >> apparent issue with .png assigmnet, commented out for now

```
